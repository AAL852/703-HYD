```python
# Function to clean and tokenize tweets
def clean_tweet(tweet):
    # Remove URLs
    tweet = re.sub(r"http\S+|www\S+|https\S+", '', tweet, flags=re.MULTILINE)
    # Remove emojis and non-ASCII characters
    tweet = re.sub(r'[^\x00-\x7F]+', '', tweet)
    # Remove punctuation
    tweet = re.sub(r'[^\w\s]', ' ', tweet)  # ç§»é™¤æ ‡ç‚¹
    # Remove extra spaces
    tweet = re.sub(r'\s+', ' ', tweet).strip()  # ç§»é™¤å¤šä½™ç©ºæ ¼
    # Lowercase
    tweet = tweet.lower()
    return tweet
```

```python
data['cleaned_tweet'] = data['tweet_text'].apply(clean_tweet)
data['tweet_text_tokens'] = data['cleaned_tweet'].apply(word_tokenize)
```

### Sentiment Analysis

```python
analyzer = SentimentIntensityAnalyzer()
```

```python
# Sentiment Analysis Function
def sentiment_analysis(text):
    scores = analyzer.polarity_scores(text)
    if scores['compound'] >= 0.05:
        return 1    # 'positive'
    elif scores['compound'] <= -0.05:
        return -1   # 'negative'
    else:
        return 0    # 'neutral'
```

```python
data['sentiment'] = data['tweet_text'].apply(sentiment_analysis)
```

### Feature Extraction

```python
#Processing hashtags and mentions columns: Replace NaN with an empty list, separate other content with '|' and convert to a list, remove unnecessary spaces, convert to lowercase for tokenization later
# --- Hashtags ---
# Extract hashtags and create hashtag count
data['hashtags_tokens'] = data['hashtags'].apply(lambda x: [tag.strip().lower() for tag in x.split('|') if tag.strip()] if isinstance(x, str) else [])
data['hashtags_count'] = data['hashtags_tokens'].apply(len)

# --- Mentions ---
# Extract mentions and create mention count
data['mentions_tokens'] = data['mentions'].apply(lambda x: [tag.strip().lower() for tag in x.split('|') if tag.strip()] if isinstance(x, str) else [])
data['mentions_count'] = data['mentions_tokens'].apply(len)

# --- User Influence ---
# Calculate user influence score ( influence = followers_count * (1 + verified) )
data['user_influence'] = data['followers_count'] * (1 + data['verified'])
data['user_influence'].fillna(0, inplace=True)
data['user_influence'] = data['user_influence'].astype(int)

# --- Timing Features ---
# Convert tweet_creation to datetime
data['tweet_creation'] = pd.to_datetime(data['tweet_creation'])
data['tweet_year'] = data['tweet_creation'].dt.year
data['tweet_month'] = data['tweet_creation'].dt.month
data['tweet_days'] = data['tweet_creation'].dt.day
data['tweet_day'] = data['tweet_creation'].dt.dayofweek
data['tweet_hour'] = data['tweet_creation'].dt.hour
data['is_weekend'] = data['tweet_creation'].dt.dayofweek >= 5
```

```python
# Initialize the SBERT model
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
# Replace NaN values with empty strings in the relevant columns
data['hashtags'] = data['hashtags'].fillna('')
data['mentions'] = data['mentions'].fillna('')
# Encode each tweet into a dense vector representation
data['hashtags_embedding'] = data['hashtags'].apply(lambda x: model.encode(x).tolist())
data['mentions_embedding'] = data['mentions'].apply(lambda x: model.encode(x).tolist())
data['tweet_text_embedding'] = data['tweet_text'].apply(lambda x: model.encode(x).tolist())
```
