```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the processed CSV file
file_path = 'data_processed_filtered1.csv'  # Replace with your file path
data = pd.read_csv(file_path)

# Create a date column for plotting (based on tweet_year, tweet_month, and tweet_days)
data['date'] = pd.to_datetime({
    'year': data['tweet_year'],
    'month': data['tweet_month'],
    'day': data['tweet_days']
})

# Plot the trend of engagement over time
plt.figure(figsize=(12, 6))
plt.plot(data['date'], data['engagement'], marker='o', linestyle='-', color='b', label='Engagement')
plt.xlabel('Date')
plt.ylabel('Engagement')
plt.title('Engagement Over Time')
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.grid(True)
plt.legend()
plt.tight_layout()

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取处理后的 CSV 文件
file_path = 'data_processed_filtered1.csv'  # 替换为您的文件路径
data = pd.read_csv(file_path)

# 过滤出 2013 到 2022 年的数据
data_filtered = data[(data['tweet_year'] >= 2013) & (data['tweet_year'] <= 2022)]

# 创建日期列以便绘图（根据 tweet_year、tweet_month 和 tweet_days）
data_filtered['date'] = pd.to_datetime({
    'year': data_filtered['tweet_year'],
    'month': data_filtered['tweet_month'],
    'day': data_filtered['tweet_days']
})

# 绘制 engagement 的变化趋势
plt.figure(figsize=(12, 6))
plt.plot(data_filtered['date'], data_filtered['engagement'], marker='o', linestyle='-', color='b', label='Engagement')
plt.xlabel('Date')
plt.ylabel('Engagement')
plt.title('Engagement Over Time (2013-2022)')
plt.xticks(rotation=45)  # 旋转 x 轴标签，方便查看日期
plt.grid(True)
plt.legend()
plt.tight_layout()

# 显示图形
plt.show()
```
